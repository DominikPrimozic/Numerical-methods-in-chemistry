```python
def matrix_vector(A,b):
    N=len(b)
    c=[0]*N
    for i in range(N):
        for j in range(N):
             c[i]+=A[i][j]*b[j]
    return c

def discrete_fourier_trnasformation(f):
    import cmath
    from math import pi
    N=len(f)
    W=cmath.exp(-1j * 2 * cmath.pi / N)
    A=[[1 for _ in range(N)] for _ in range(N)]
    for i in range(N):
        for j in range(N):
            A[i][j]=W**(i*j)
    return matrix_vector(A,f)
```

```python
f=[8,4,8,0]
discrete_fourier_trnasformation(f)
```
