```python
def norm(vec):
    s=0
    for i in vec:
        s+=i*i
    return s**0.5
def vector_difference(v1,v2):
    assert len(v1)==len(v2)
    newvec=v1.copy()
    for i in range(len(v2)):
        newvec[i]-=v2[i]
    return newvec
    
class Gauss_Seidel_solver():
    #Solving Ax=b
    def __init__(self,A,b):
        self.run(A,b)

    def run(self,A,b):
        x=[0]*len(A)
        for k in range(1000):
            x0=x.copy()
            for i in range(len(A)):
                up=0
                down=0
                for j in range(i):
                    down+=(-A[i][j]*x[j])
                for j in range(i+1,len(A)):
                    up+=(-A[i][j]*x0[j]) 
                x[i]=(b[i] + up + down)/A[i][i]
            if norm(vector_difference(x,x0))<1e-6: 
                self.solution=x
                return 
        self.solution=None
        return 
```

```python
A=[[2,3],[4,9]]
b=[6,15]
```

```python

solver = Gauss_Seidel_solver(A, b)
print(solver.solution)
```

```python
class finite_difference():

    def __init__(self,h,c,boundary,T0,Tn):
        self.N=len(boundary)
        self.constructA(h,c)
        self.constructb(h,c,boundary,T0,Tn)
        solver=Gauss_Seidel_solver(self.A,self.b)
        self.sol=solver.solution
    def constructA(self,h,c):
        A=[]
        for _ in range(self.N):
            A.append([0]*self.N)
        for i in range(self.N):
            A[i][i]=-(h*h*c+2)
            if i<self.N-1:
                A[i][i+1]=1
            if i>0:
                A[i][i-1]=1 
        self.A=A
    def constructb(self,h,c,boundary,T0,Tn):   
        b=boundary.copy()
        for i in range(len(b)):
            b[i]*=-h*h*c
        b[0]-=T0
        b[self.N-1]-=Tn
        self.b=b
```
