```python
def vector_norm(v1,v2):
        return sum((v_1-v_2)**2 for v_1,v_2 in zip(v1,v2)) ** 0.5
    
def jacobi_iteration(A,b,x0):
    n=len(A)
    x=x0.copy()
    for _ in range(1000):
        for i in range(n):
            s=0
            for j in range(n):
                if j != i:
                    s+=A[i][j]*x0[j]
            x[i]=(b[i]-s)/A[i][i]
        if vector_norm(x,x0)<1e-6:
            print(_)
            return x
        x0=x.copy()
    return 
    
```

```python
A = [
    [4, 1, 2],
    [3, 5, 1],
    [1, 1, 3]
]
b = [9, 7, 3]
x0 = [0, 0, 0] 
jacobi_iteration(A, b, x0)
```

```python
def gauss_seidel_iteration(A,b,x0):
    n=len(A)
    x=x0.copy()
    for _ in range(1000):
        for i in range(n):
            s=0
            for j in range(n):
                if j != i:
                    s+=A[i][j]*x[j]
            x[i]=(b[i]-s)/A[i][i]
        if vector_norm(x,x0)<1e-6:
            print(_)
            return x
        x0=x.copy()
    return 
```
